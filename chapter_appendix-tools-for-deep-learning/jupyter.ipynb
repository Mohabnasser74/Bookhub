```python
import pandas as pd
```

```python
training_data = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
training_data.isna().sum()
```

```python
x_train = training_data.drop("housing_median_age", axis=1, inplace=False)
y_train = training_data['housing_median_age']
```

```python
testing_data = pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
testing_data.isna().sum()
```

```python
x_test= testing_data.drop("housing_median_age", axis=1)
y_test = testing_data["housing_median_age"]
```

```python
print(training_data.shape)
print(testing_data.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinearRegressionModel = LinearRegression(fit_intercept=True, copy_X=True, n_jobs=-1)
```

```python
%%time
LinearRegressionModel.fit(x_train, y_train)
```

```python
LinearRegressionModel.score(x_train, y_train)
```

```python
y_pred = LinearRegressionModel.predict(x_test)
```

```python
print("predict y: ", y_pred)
print("test x: ", x_test)
```
